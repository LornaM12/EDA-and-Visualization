```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
haberdata = pd.read_csv('haberman.csv')
```

```python
print(haberdata.head())
```

```python
haberdata['status'].value_counts()
```

```python
haberdata.plot(kind = 'scatter' , x= 'age' , y= 'nodes')
plt.title('Scatter Plot of Age against nodes')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(haberdata,hue='status',height=4) \
.map(plt.scatter, 'age', 'nodes') \
.add_legend()
plt.title('Scatter plot of age against nodes')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(haberdata,hue='status', height=3)
plt.show()
```

```python
sns.FacetGrid(haberdata,hue='status', height=5) \
.map(sns.histplot, 'age') \
.add_legend()
plt.title('Age represantation')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(haberdata, x='age', hue='status', height=5)
plt.title('Age representation')
plt.show()
```

```python
sns.boxplot(x='status', y='nodes', data=haberdata)
plt.legend
plt.title('Box Plot')
plt.show()
```
